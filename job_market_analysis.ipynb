# Data Cleaning

```python
# Remove Duplicate
df = df.drop_duplicates()
df
```

```python
# Handling Missing Values
df = df.copy()
df['salary_usd'] = df['salary_usd'].fillna(df['salary_usd'].median())
df
```

```python
df['location'] = df['location'].fillna('Unknown')
```

```python
#  Cleaning Text Columns
df['job_title'] = df['job_title'].astype(str).str.strip()

df['industry'] = df['industry'].astype(str).str.strip()

df['location'] = df['location'].astype(str).str.strip()

df['skills_required'] = df['skills_required'].astype(str).str.strip()

df['remote_option'] = df['remote_option'].astype(str).str.strip()

df['company_size'] = df['company_size'].astype(str).str.strip()
```

```python
# Converting Date Column
df['posting_date'] = pd.to_datetime(df['posting_date'], dayfirst=True)
```

```python
# Extracting Month and Month Name
df['month'] = df['posting_date'].dt.month
df['month_name'] = df['posting_date'].dt.month_name()
```

```python
#  Data Transformation for Export
df['skills_required'] = df['skills_required'].str.replace(',', ';')
```

# Exporting Cleaned Dataset

```python
import csv
df.to_csv("cleaned_job_data.csv", index=False, quoting=csv.QUOTE_ALL)
```
